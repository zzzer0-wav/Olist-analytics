### Part 3. Python: Simple forecast and recommendations

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
url = 'https://raw.githubusercontent.com/zzzer0-wav/Olist-analytics/refs/heads/main/data/main_sales_reg(excel).csv'
df = pd.read_csv(url)
RANDOM_STATE = 42
```

```python
df.head()
```

```python
df.info()
```

```python
miss = df.isna().sum()
miss_pct = round(df.isna().mean() * 100, 2)
missing_table = pd.DataFrame({
    'type':df.dtypes.astype(str),
    'missing': miss,
    'missing_pct':miss_pct
})

missing_table
```

```python
df = df.sort_values('ym').reset_index(drop=True)
print('Month', len(df))
df.head()
```

```python
df = df[(df.ym >= '2017-01') & (df.ym <= '2018-08')].reset_index(drop=True)
print('Month (in process): ', len(df))
display(df.head(2))
df.tail(2)
```

```python
pivot = df.pivot_table(
    values='price',
    index='ym',
    aggfunc='sum'
)

pivot
```

```python
plt.figure(figsize=(10, 6))
plt.plot(pivot.index, pivot.price, marker='o')
plt.xticks(rotation=45, ha='right')
plt.title('Monthly revenue (stable period)')
plt.tight_layout()
plt.show()
```

```python
df_new = pd.DataFrame({
    't':list(range(len(pivot))),
    'ym':list(pivot.index),
    'revenue':list(pivot.price)
})

df_new
```

```python
list(pivot.index)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

X = df_new[['t']]
y = df_new['revenue']

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.1,
    random_state=RANDOM_STATE
)

model = LinearRegression().fit(X_train, y_train)

next_index = len(df_new)
y_pred = model.predict(X_test)

forecast = model.predict(pd.DataFrame({
    't':[next_index]
}))[0]
print(f'Predict revenue for next month: R${forecast:,.0f}')
print(f'Average growth for month: R${model.coef_[0]:,.0f}')
print('-'*35)
print(f'MAE: {mean_absolute_error(y_test, y_pred):,.0f}')
print(f'RMSE: R${np.sqrt(mean_squared_error(y_test, y_pred)):,.0f}')
print(f'Predict accuracy (R2):{r2_score(y_test, y_pred):,.0f}')
```

```python
df_new['month'] = df_new['ym'].str[-2:]
seasons = df_new.groupby('month')['revenue'].mean().round(0)
print('Average growth for number month:')
print(seasons.sort_values(ascending=False))
```
